```python
# Capture the list of dog breed name from dogfacts.csv file.

import pandas as pd
df= pd.read_csv('dogfacts.csv')
dog_list = df[df.columns[0]].tolist()

# replace space with '-' to fit the petplace website format
dogs_list = [dog.replace(' ', '-').lower() for dog in dog_list]
```

```python
import requests
from bs4 import BeautifulSoup
import json


def scrape_dog_descriptions(dogs_list):
    dog_descriptions = {}
    for dog in dogs_list:
        url = f"https://www.petplace.com/article/breed/{dog}"
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            div = soup.find('div', {'class': 'blade'})
            if div:
                dog_descriptions[dog] = div.text.strip()
            else:
                dog_descriptions[dog] = 'No description found'
        else:
            dog_descriptions[dog] = 'Page not found'
    return dog_descriptions

dog_descriptions = scrape_dog_descriptions(dogs_list)
```
