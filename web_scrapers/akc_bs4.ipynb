```python
from bs4 import BeautifulSoup
import requests, json
import pandas as pd

def get_soup(url):
    r = requests.get(url)
    data = r.text
    soup = BeautifulSoup(data,'html')
    return soup

def get_dog_info(dog_breed):
    url = 'https://www.akc.org/dog-breeds/{}'.format(dog_breed)
    soup = get_soup(url)
    json_data = json.loads(soup.select('[type="application/ld+json"]')[2].text)

    dog_name = {'name': json_data['name']}
    dog_description = {'description': json_data['description']}
    dog_measure = json_data['mainEntity']['csvw:tableSchema']['csvw:columns']

    dog_measure_data = extract_measure_data(dog_measure)

    dog_info = {}
    dog_info.update(dog_name)
    dog_info.update(dog_description)
    dog_info.update(dog_measure_data)

    return dog_info

def extract_measure_data(dog_measure):
    dog_measure_data = {}
    for r in dog_measure:
        key, value = r['csvw:cells'][0]['csvw:value'].split(': ')
        dog_measure_data[key] = value
    return dog_measure_data

def scrape_all_breeds(dogs_list):
    all_dog_info = []
    for dog_breed in dogs_list:
        print(f"Fetching info for {dog_breed}...")
        try:
            dog_info = get_dog_info(dog_breed)
            all_dog_info.append(dog_info)
        except Exception as e:
            print(f"Could not fetch info for {dog_breed}: {str(e)}")
    return all_dog_info
```
