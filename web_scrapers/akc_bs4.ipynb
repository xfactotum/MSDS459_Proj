```python
from bs4 import BeautifulSoup
import requests, json
import pandas as pd
```

```python
url = 'https://www.akc.org/dog-breeds/Affenpinscher/'
r = requests.get(url)
data = r.text
soup = BeautifulSoup(data,'html')
```

```python
print(soup.prettify())
```

```python
# this is where to get the dog breed summary
json.loads(soup.select('[type="application/ld+json"]')[2].text)
```

```python
dog_name = {'name':json.loads(soup.select('[type="application/ld+json"]')[2].text)['name']}
dog_description = {'description':json.loads(soup.select('[type="application/ld+json"]')[2].text)['description']}
```

```python
dog_measure = json.loads(soup.select('[type="application/ld+json"]')[2].text)['mainEntity']['csvw:tableSchema']['csvw:columns']
dog_measure_data = {}
for r in dog_measure:
#     print(r['csvw:cells'][0]['csvw:value'])
    key, value = r['csvw:cells'][0]['csvw:value'].split(': ')
    dog_measure_data[key] = value
```

```python
dog_info = {}
dog_info.update(dog_name)
dog_info.update(dog_description)
dog_info.update(dog_measure_data)
```
